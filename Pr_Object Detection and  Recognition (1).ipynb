# <center> Object Detection & Recognition <center>

Image classification involves assigning a class label to an image, whereas object 
localization involves drawing a bounding box around one or more objects in an image. 
Object detection is more challenging and combines these two tasks and draws a bounding 
box around each object of interest in the image and assigns them a class label. Together, 
all of these problems are referred to as object recognition.

### Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

```python
config_file=r"C:\Users\HP\Desktop\ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt"
frozen_model=r"C:\Users\HP\Desktop\ssd_mobilenet_v3_large_coco_2020_01_14\frozen_inference_graph.pb"
```

```python
model=cv2.dnn_DetectionModel(frozen_model,config_file)
```

```python
classLabels=[]
file_name=r"C:\Users\HP\Desktop\Labels.txt"
with open(file_name,'rt') as fpt:
    classLabels=fpt.read().rstrip('\n').split('\n')
    
```

```python
print(classLabels)
```

```python
print(len(classLabels))
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean((127.5,127.5,127.5))
model.setInputSwapRB(True)
```

### Read an Image

```python
img=cv2.imread(r"C:\Users\HP\Downloads\pexels-andrea-piacquadio-774909.jpg")
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
ClassIndex,confidence,bbox=model.detect(img,confThreshold=0.5)
```

```python
print(ClassIndex)
```

```python
font_scale=30
font=cv2.FONT_HERSHEY_PLAIN
for ClassInd,conf,boxes in zip(ClassIndex,confidence,bbox):
    cv2.rectangle(img,boxes,(255,0,0),2)
    cv2.putText(img,classLabels[ClassInd-1],(boxes[0]+10,boxes[1]+40),font,fontScale=font_scale,color=(0,255,0),thickness=30)
```
